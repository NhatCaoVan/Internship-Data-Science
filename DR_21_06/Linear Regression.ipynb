```python
#Trên numpy chúng ta có thể tính được ma trận hiệp phương sai giữa x, y như sau:
import numpy as np

# Khởi tạo các vector x, y ngẫu nhiên có độ dài 100
x = np.random.randn(100)
y = np.random.randn(100)

# print(x)
# print(y)

# Ma trận hiệp phương sai
np.cov(x, y)
```

```python
# Phương sai (Variance)
# Để tính phương sai của véc tơ x trên numpy ta dùng hàm np.var().
# Phương sai của x
np.var(x, ddof=1)

# Hoặc 
x.var(ddof=1)
```

```python
# Độ lệch chuẩn (Standard Deviation)
# Độ lệch chuẩn được tính theo hàm np.std()
# Phương sai của x
np.std(x, ddof=1)

# Hoặc
x.std(ddof=1)
```

```python
# Hệ số tương quan (Correlation Coefficient)
# Ma trận hệ số tương quan giữa hai véc tơ x, y được tính theo hàm np.corrcoef()
# Hệ số tương quan giữa x và y
x = np.random.randn(100)
y = np.random.randn(100)
np.corrcoef(x, y)
```

```python
# Trong python khi làm việc với phân phối xác suất và thống kê chúng ta sẽ chủ yếu sử dụng
# package scipy. Hàm pdf của phân phối chuẩn có thể được tính dựa vào hàm norm.pdf() của scipy.
import numpy as np
from scipy.stats import norm

x = np.array([1, 2, 3, 4, 5])
# Tính pdf cho véc tơ x theo phân phối chuẩn hoá
norm.pdf(x)
```

```python
# Hàm phân phối xác suất tích luỹ (cdf)
# Để tính cdf cho phân phối chuẩn chúng ta dùng hàm cdf()
import numpy as np
from scipy.stats import norm

# Tính cdf(x)= p(X<=x)
x = 2.1
p = norm.cdf(x)
p
```

```python
# Từ xác xuất tích luỹ CDF(x) ta có thể tính ra được giá trị của x dựa vào 
# hàm ppd() (Percent Point Function, hàm ngược của CDF). Đây chính là gía trị tới hạn của phân phối xác suất.
# Tính giá trị x dựa vào xác suất p=cdf(x)
norm.ppf(p)
```

```python
# Phân phối chuẩn (Gaussian distribution)
import torch
import matplotlib.pyplot as plt
import seaborn as sns

def _plot(x, title, figsize=(10, 6)):
  plt.figure(figsize=figsize)
  sns.kdeplot(x)
  plt.xlabel("Value of x")
  plt.title(title)

x = torch.randn(1000)
_plot(x, "Standard Normal Distribution")
```

```python
x = torch.normal(mean=1, std=2, size=(1000,))
_plot(x, "Normal Distribution")
```

```python
# Phân phối đều (uniform distribution)
x = torch.rand(1000)
_plot(x, "Uniform Distribution")
```

```python
#  Phân phối Poisson
# Phân phối Poisson trên pytorch được khởi tạo thông qua hàm torch.poisson(). 
rates = torch.rand(1000)
x = torch.poisson(rates)
_plot(x, "Poisson Distribution")
```
